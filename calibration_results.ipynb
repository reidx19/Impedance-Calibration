```python
import pickle
from pathlib import Path
import pandas as pd
```

```python
fp = Path.home() / "Documents/GitHub/Impedance-Calibration"
with (fp/"calibration_results.pkl").open('rb') as fh:
    results = pickle.load(fh)
```

```python
new_results = {key.parts[-1].split('.csv')[0]:items for key, items in results.items()}
new_results.keys()
```

```python
df = pd.DataFrame.from_dict(new_results,orient='index',columns=['array','overlap_prop'])#,columns=)
df[['not beltline','not bike infra','2+ lanes','> 30 mph']] = df['array'].apply(pd.Series)
df.drop(columns=['array'],inplace=True)
df['overlap_prop'] = df['overlap_prop'].abs()
```

```python
df.sort_values('overlap_prop',ascending=False).round(2)
```
